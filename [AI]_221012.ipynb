```python
# 유임 승차 비율
import csv

fp = open("/content/gdrive/MyDrive/Colab/subwayfee.csv", encoding='cp949')
data = csv.reader(fp)

# rate = 유임승차인원 / 무임승차인원
rate = 0
max_rate = 0
next(data)

for row in data:
  for i in range(4,8):
    row[i] = int(row[i].replace(",",""))

  # 10만 이상
  if row[6]!=0 and (row[4]+row[6]) >100000:
    rate = row[4]/(row[4]+row[6])
  # 0.94 이상
  if rate > 0.94:
    print(row, " " ,round(rate,2))

fp.close()
```

```python
# 홍대 입구 최고 유임 승차 비율
import csv

fp = open("/content/gdrive/MyDrive/Colab/subwayfee.csv", encoding='cp949')
data = csv.reader(fp)

rate=0
max_rate = 0
max_st =''
next(data)

for row in data:
  for i in range(4,8):
    row[i] = int(row[i].replace(",",""))
  if '홍대' in row[3]:
    if row[6]!=0 and (row[4]+row[6]) >100000:
      rate = row[4]/(row[4]+row[6])
    if rate > max_rate:
      max_rate = rate
      max_st = row[3]+ " " +row[1]

print(max_st,":",round(max_rate*100, 2))

fp.close()
```

# map(f, iter)
- map(바꿀 자료형/함수, 바꿀 내용)

```python
import csv
import matplotlib.pyplot as plt

uri = "/content/gdrive/MyDrive/Colab/subwaytime.csv"

fp = open(uri, encoding="cp949")
data = csv.reader(fp)

result=[]
next(data)
next(data)

for row in data:
  row[4:] = map(int,row[4:])
  result.append(row[10])

fp.close()

plt.style.use("ggplot")
plt.title("subwaytime")
plt.bar(range(len(result)),result)
plt.show()
```

```python
# 7~9시 승차 인원 최대인 역과 사람 수는?
import csv
import matplotlib.pyplot as plt

uri = "/content/gdrive/MyDrive/Colab/subwaytime.csv"

fp = open(uri, encoding="cp949")
data = csv.reader(fp)

max_data=0
max_st=''
next(data)
next(data)

for row in data:
  row[4:] = map(int,row[4:])
  if sum(row[10:15:2]) > max_data:
    max_data = sum(row[10:15:2])
    max_st = row[3] +" "+row[1]

fp.close()
print("7~9시 승차 인원 최대 :",max_st, ":",max_data)
```

```python
# 23시에 가장 많이 타고 내리는 역과 사람 수는?

import csv
import matplotlib.pyplot as plt

uri = "/content/gdrive/MyDrive/Colab/subwaytime.csv"

fp = open(uri, encoding="cp949")
data = csv.reader(fp)

max_data=0 # 최대 인원수
max_st=''
t=23
next(data)
next(data)

for row in data:
  # str > int
  row[4:] = map(int,row[4:])
  m= row[4+(t-4)*2] # 승차시간 t[4,5,6, ...] -> 인덱스 m[4,6,8, ...] 관계
  
  # 23시에 최대 인원수와 역 찾기
  if m > max_data:
    max_data = m
    max_st = row[3] +" "+row[1]

fp.close()
print("23시 가장 많이 타고 내리는 곳 :",max_st, ":",max_data)
```
