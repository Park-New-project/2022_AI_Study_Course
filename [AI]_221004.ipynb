```python
from bs4 import BeautifulSoup

html = """
<html><body>
  <h1>스크레이핑 연습</h1>
  <p>웹 페이지를 분석하기</p>
  <p>데이터를 정제하기</p>
</body></html>
"""

#1. 객체 생성
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
# 2. 원하는 요소 접근
h1 = soup.html.body.h1
h1
```

```python
p1 = soup.html.body.p
p1
```

```python
p2 = p1.next_sibling.next_sibling
p2
```

```python
# 3. 데이터 추출
print("h1 :",h1.string)
print("p1 :",p1.string)
print("p2 :",p2.string)
```

# .find()

```python
from bs4 import BeautifulSoup

html = """
<html><body>
  <h1 id="title">스크레이핑 연습</h1>
  <p id="subTitle">웹 페이지를 분석하기</p>
  <p>데이터를 정제하기</p>
</body></html>
"""

soup = BeautifulSoup(html, 'html.parser')
soup

title = soup.find(id="title")
subTitle = soup.find(id="subTitle")

print("title :", title.string)
print("subTitle :", subTitle.string)
```

- < ul > : 순서가 없는 리스트

```python
from bs4 import BeautifulSoup

html = """
<html><body>
  <ul>
    <li><a href="http://www.naver.com">네이버</a><li>
    <li><a href="http://www.daum.net">다음</a><li>
  </ul>
  
</body></html>
"""
soup = BeautifulSoup(html, 'html.parser')

links = soup.find_all('a')
links
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res = req.urlopen(url)
print(res)
```

```python
soup = BeautifulSoup(res, 'html.parser')
soup
```

```python
title = soup.find('title').string
print(title)
```

```python
wf = soup.find('wf').string
print(wf)
```
