```python
# 완성된 코드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/gdrive/MyDrive/Colab/age.csv',encoding='cp949', index_col=0, thousands = ',')
df = df.div(df['2019년02월_계_총인구수'], axis=0)

del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']

name= "신도림"

a = df.index.str.contains(name)
df2 = df[a]

plt.style.use("ggplot")
plt.rc('font', family='NanumBarunGothic', size=12)
df.loc[np.power(df.sub(df2.iloc[0], axis=1) ,2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```

```python
# 완성된 코드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/gdrive/MyDrive/Colab/age.csv',encoding='cp949', index_col=0, thousands = ',')
df = df.div(df['2019년02월_계_총인구수'], axis=0)

del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']

name= input("원하는 지역을 입력 : ")

a = df.index.str.contains(name)
df2 = df[a]

plt.style.use("ggplot")
plt.rc('font', family='NanumBarunGothic', size=12)
df.loc[np.power(df.sub(df2.iloc[0], axis=1) ,2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```
