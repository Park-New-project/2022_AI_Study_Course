```python
data = boston.data.astype('float32')
data.dtype
```

```python
boston.target.dtype
```

```python
target = boston.target.astype('float32')
target.dtype
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2)

print(len(X_train),"" ,len(y_test))
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.3)

print(len(X_train),"" ,len(y_valid))
```

```python
import boto3
import sagemaker
from sagemaker import get_execution_role
```

```python
boto3.Session()
```

```python
region = boto3.Session().region_name

print(region)
print(type(region))
```

```python
role = get_execution_role()

print(role)
print(type(role))
```

```python
bucket = "sagemaker-pmj-20221115"


print(bucket)
print(role)
print(region)
```

```python
output_location = "s3://{}/LinearLearnerExample.output".format(bucket)
print(output_location)
```

```python
#모델의 표현 방법 : sagemaker 내장알고리즘
linear = sagemaker.LinearLearner(
    role=role,
    instance_count =1,
    instance_type = 'ml.m5.large',
    output_path = output_location,
    predictor_type='regressor',
    early_stopping_patience=100
)
linear
```

```python
train = linear.record_set(X_train, labels=y_train, channel='train')
print(train)
```

```python
validation = linear.record_set(X_valid, labels=y_valid, channel='validation')
print(validation)
```

```python
test = linear.record_set(X_test, labels=y_test, channel='test')
print(test)
```

```python
print(train)
print(validation)
print(test)
```

```python
linear.fit([train, validation, test], 
           mini_batch_size=10,
          job_name='LinearLearner-Boston-housing-price-01')
```
