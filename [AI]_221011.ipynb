```python
# 제주도의 성별 비율 바형 시각화
import csv
import matplotlib.pyplot as plt

fp = open("/content/gdrive/MyDrive/Colab/gender.csv", encoding='cp949')
data = csv.reader(fp)

male=[]
female=[]

for row in data:
  if '제주' in row[0]:
    for i in row[3:104]:
      male.append(-int(i.replace(',','')))
      
    for i in row[106:]:
      female.append(int(i.replace(',','')))
    break
fp.close()

plt.style.use("ggplot")
plt.title("제주")
plt.barh(range(101), male, label="남성")
plt.barh(range(101), female, label="여성")
plt.legend()
plt.show()
```

```python
# 제주도의 성별 비율 파이형 시각화
import csv
import matplotlib.pyplot as plt

fp = open("/content/gdrive/MyDrive/Colab/gender.csv", encoding='cp949')
data = csv.reader(fp)

size=[]

for row in data:
  if '제주' in row[0]:
    male_total = 0
    female_total = 0
    for i in range(101):
      male_total += int(row[i+3].replace(',',''))
      female_total += int(row[i+106].replace(',',''))
    break
fp.close()

size.append(male_total)
size.append(female_total)

color = ['crimson', 'darkcyan']

plt.rc('font',family='NanumBarunGothic', size=12)
plt.axis('equal')
plt.title("제주")
plt.pie(size, colors = color, labels = ['남','여'], autopct="%.1f%%", startangle = 90)

plt.legend()
plt.show()
```

```python
# 신도림의 성별 비율 파이형 시각화
import csv
import matplotlib.pyplot as plt

fp = open("/content/gdrive/MyDrive/Colab/gender.csv", encoding='cp949')
data = csv.reader(fp)

size=[]
area = input("성별 비율을 알고 싶은 지역 입력 : ")

for row in data:
  if area in row[0]:
    male_total = 0
    female_total = 0
    for i in range(101):
      male_total += int(row[i+3].replace(',',''))
      female_total += int(row[i+106].replace(',',''))
    break
fp.close()

size.append(male_total)
size.append(female_total)

color = ['crimson', 'darkcyan']

plt.rc('font',family='NanumBarunGothic', size=12)
plt.axis('equal')
plt.title(area)
plt.pie(size, colors = color, labels = ['남','여'], autopct="%.1f%%", startangle = 90)

plt.legend()
plt.show()
```

```python
# 제주도의 성별 비율 바형 시각화
import csv
import matplotlib.pyplot as plt

fp = open("/content/gdrive/MyDrive/Colab/gender.csv", encoding='cp949')
data = csv.reader(fp)

male=[]
female=[]
area = input("성별 비율을 알고 싶은 지역 입력 : ")

for row in data:
  if area in row[0]:
    for i in range(3,104):
      male.append(int(row[i].replace(',','')))
      female.append(int(row[i+103].replace(',','')))
    break
fp.close()

plt.rc('font',family='NanumBarunGothic', size=12)
plt.title(area)
plt.scatter(male, female, c=range(101), cmap='jet')
plt.colorbar()

plt.plot(range(max(male)), range(max(male)), "g")

plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
