```python
# 기상청 RSS 데이터 추출
import urllib.request
import urllib.parse
```

```python
rssUrl = "https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
values = {
    'stnid' : '108'
}
type(values)
```

```python
params = urllib.parse.urlencode(values)
url = rssUrl + "?" + params
print(url)
```

```python
data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")

print(text)
```

# BeautifulSoup
- XML, HTML을 분석해주는 라이브러리

```python
from bs4 import BeautifulSoup

html = """
<html><body>
  <h1>스크레이핑 연습</h1>
  <p>웹 페이지를 분석하기</p>
  <p>데이터를 정제하기</p>
</body></html>
"""

# 1. 객체 생성
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
# 2. 원하는 요소 접근
h1 = soup.html.body.h1
h1
```

```python
p1 = soup.html.body.p
p1
```
