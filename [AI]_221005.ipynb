```python
# 서울이 가장 더웠던 날짜와 기온?
# 최고 기온 데이터 사용하기 
import csv

f = open("seoul.csv", encoding='cp949')
data = csv.reader(f, delimiter=",")

next(data) # 헤더 제거
max_temp = -999.0

for row in data:
  if row[4] =='':
    row[4] = -999 # 데이터 공백 -> -999 로 초기화 
  
  # str > float 형변환
  row[4] = float(row[4]) 

  # 최고기온 데이터 변수저장
  if max_temp < row[4]:
     max_temp = row[4]
     max_date = row[0]

f.close()
print("기상청 관측 아래 서울이 가장 더웠던 날짜는? 날짜 :",max_date, "이고, 기온:",max_temp)
```

```python
# [직접]서울이 가장 추웠던 날짜와 기온?
# 최저 기온 데이터 사용하기 
import csv

f = open("seoul.csv", encoding='cp949')
data = csv.reader(f, delimiter=",")

next(data)
min_temp = 999.0

for row in data:
  if row[3] =='':
    row[3] = 999

  row[3] = float(row[3]) 

  if min_temp > row[3]:
     min_temp = row[3]
     min_date = row[0]

f.close()
print("기상청 관측 아래 서울이 가장 더웠던 날짜는? 날짜 :",min_date, "이고, 기온:",min_temp)
```
